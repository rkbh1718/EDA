```python
# dataset link : https://www.kaggle.com/sdolezel/black-friday/version/1?select=train.csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# importing blackfriday_train dataset
df_train = pd.read_csv('blackfriday_train.csv')
df_train.head()
```

```python
# importing test dataset
df_test = pd.read_csv('test.csv')
df_test.head()
```

```python
# merging both dataset

#pd.merge()

df = df_train.append(df_test)
df.head()
```

```python
pd.get_dummies(df['Gender'])
```

```python
## handling calogirical feature Gender
df['Gender']=df['Gender'].map({'F':0 , 'M':1})
df.head()
```

```python
## handle catogirical feature AGE
df['Age'].uniqueque()
```

```python
#pd.get_dummies(df['Age'],drop_first= True)
df['Age']= df['Age'].map({'0-17':1, '18-25':2, '26-35':3, '36-45':4, '46-50':5, '51-55':6, '55+':7})
```

```python
df.head()
```

```python
## Second technique

# Import label encoder
from sklearn import preprocessing
 
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
 
# Encode labels in column 'species'.
df['Age']= label_encoder.fit_transform(df['Age'])
 
df['Age'].unique()
# but it will start with 0 to 6 but in above i have put from 1 to 7
```

```python
## fixing categorical City_Category

df_city = pd.get_dummies(df['City_Category'], drop_first=True)  ## here , drop_first=True means first A drop
```
